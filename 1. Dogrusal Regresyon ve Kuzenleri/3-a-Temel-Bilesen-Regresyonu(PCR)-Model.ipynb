```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
hit = pd.read_csv("Hitters.csv")
```

```python
df = hit.copy()
```

```python
# hızlı bir ön işleme işlemi yapmamız gerekiyor.
# örneğin, buradaki kategorik değişkenleri 'dummy' değişken formatına çevirebiliriz.

dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])
dms.head()
```

```python
dms.shape
```

```python
y = df["Salary"]
```

```python
# bağımlı değişkeni ve kategorik değişkenlerin ilk halini çıkarmak için;

X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis = 1).astype("float64")
```

```python
X_.head()
```

```python
X_.shape
```

```python
# BAĞIMSIZ DEĞİŞKENLERİMİZ

X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis = 1)
```

```python
X.head()
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

print("X_train dim  :", X_train.shape)
print("y_train dim  :", y_train.shape)
print("X_test dim   :", X_test.shape)
print("y_test dim   :", y_test.shape)

training = df.copy()

print("training dim :", training.shape)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

pca = PCA()
```

```python
X_reduced_train = pca.fit_transform(scale(X_train))
```

```python
X_reduced_train[0:1, :]
```

```python
X_reduced_train[0:1, 0:2]
```

```python
X_reduced_train[0:1, 0:1]
```

```python
# cumilative sum

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)
```

```python
# birinci değişkenin, veri setinde bulunan toplam değişkenliğin %38.18'ini açıklaması.

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)[0:1]
```

```python
# birinci ve ikinci değişkenlerin, veri setinde bulunan toplam değişkenliğin %59.88'ini açıklaması.

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)[1:2]
```

```python
# 1., 2., ve 3. değişkenlerin, veri setinde bulunan toplam değişkenliğin %70.88'ini açıklaması.

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)[2:3]
```

```python
# bileşen sayısı arttıkça orjinal veri setinin %100'ü de açıklanmış oluyor.

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)[18:19]
```

```python
# temel bileşen analizinin temel mantığı da budur.
# ben bütün değişkenler ile uğraşacağıma
# giderim ilk 5 değişken ile uğraşıp çıkarım yaparım.
# hepsi ile uğraşmama gerek yok.
# örnek : %84.18'lik açıklama payı bana yeterli gelebilir.
# amacımız veri setini daha az bileşene indirmektir.
# 19 bileşenle temsil edersek : % 100
# 5 bileşenle temsil edersek : % 84.18

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)[4:5]
```
