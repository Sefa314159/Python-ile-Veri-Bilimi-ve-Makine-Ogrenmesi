```python
# hƒ±zlƒ± bir √∂n i≈üleme i≈ülemi yapmamƒ±z gerekiyor.
# √∂rneƒüin, buradaki kategorik deƒüi≈ükenleri 'dummy' deƒüi≈üken formatƒ±na √ßevirebiliriz.

dms = pd.get_dummies(df[["League", "Division", "NewLeague"]])
dms.head()
```

```python
dms.shape
```

```python
y = df["Salary"]
```

```python
# baƒüƒ±mlƒ± deƒüi≈ükeni ve kategorik deƒüi≈ükenlerin ilk halini √ßƒ±karmak i√ßin;

X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis = 1).astype("float64")
```

```python
X_.head()
```

```python
X_.shape
```

```python
# BAƒûIMSIZ DEƒûƒ∞≈ûKENLERƒ∞Mƒ∞Z

X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis = 1)
```

```python
X.head()
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

print("X_train dim  :", X_train.shape)
print("y_train dim  :", y_train.shape)
print("X_test dim   :", X_test.shape)
print("y_test dim   :", y_test.shape)

training = df.copy()

print("training dim :", training.shape)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

pca = PCA()
```

```python
X_reduced_train = pca.fit_transform(scale(X_train))
```

```python
X_reduced_train[0:1, :]
```

```python
# cumilative sum

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)
```

```python
# birinci deƒüi≈ükenin, veri setinde bulunan toplam deƒüi≈ükenliƒüin %38.18'ini a√ßƒ±klamasƒ±.

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)[0:1]
```

```python
# birinci ve ikinci deƒüi≈ükenlerin, veri setinde bulunan toplam deƒüi≈ükenliƒüin %59.88'ini a√ßƒ±klamasƒ±.

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)[1:2]
```

```python
# 1., 2., ve 3. deƒüi≈ükenlerin, veri setinde bulunan toplam deƒüi≈ükenliƒüin %70.88'ini a√ßƒ±klamasƒ±.

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)[2:3]
```

```python
# bile≈üen sayƒ±sƒ± arttƒ±k√ßa orjinal veri setinin %100'√º de a√ßƒ±klanmƒ±≈ü oluyor.

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)[18:19]
```

```python
# temel bile≈üen analizinin temel mantƒ±ƒüƒ± da budur.
# ben b√ºt√ºn deƒüi≈ükenler ile uƒüra≈üacaƒüƒ±ma
# giderim ilk 5 deƒüi≈üken ile uƒüra≈üƒ±p √ßƒ±karƒ±m yaparƒ±m.
# hepsi ile uƒüra≈ümama gerek yok.
# √∂rnek : %84.18'lik a√ßƒ±klama payƒ± bana yeterli gelebilir.
# amacƒ±mƒ±z veri setini daha az bile≈üene indirmektir.
# 19 bile≈üenle temsil edersek : % 100
# 5 bile≈üenle temsil edersek : % 84.18

np.cumsum(np.round(pca.explained_variance_ratio_, decimals = 4) * 100)[4:5]
```
